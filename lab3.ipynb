```python
# функция оконного преобразования Фурье


def stft(x, frame_length, hop_length):
    frames = []

    for i in range(0, len(x), hop_length):
        frame = x[i : i + frame_length]

        if len(frame) < frame_length:
            frame = np.pad(frame, (0, frame_length - len(frame)))

        frame = frame * np.hanning(frame_length)
        spec = np.fft.rfft(frame)
        amp = np.abs(spec)

        frames.append(amp)
    
    return np.array(frames)

# функция вычисления спектрального центроида

def sc(S, fs):
    N = len(S[0])
    SC = []

    for spectrum in S:
        numer, denumer = 0, 0
        for k in range(N):
            f_k = k*fs/N
            numer += spectrum[k]*f_k
            denumer += spectrum[k]
        SC.append(numer/denumer)

    return SC
```
