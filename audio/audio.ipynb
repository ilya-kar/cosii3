```python
import torchaudio
from df.enhance import enhance, init_df
from nisqa.NISQA_model import nisqaModel

snr_values = [0, 5, 10, 15, 20, 25]
```

```python
model, df_state, _ = init_df(model_base_dir="DeepFilterNet2")

for snr_db in snr_values:
    audio, sr = torchaudio.load(f"../audiofiles/noisy_{snr_db}db.wav")
    enhanced = enhance(model, df_state, audio)
    torchaudio.save(f"../audiofiles/denoised_{snr_db}db.wav", enhanced, sr)
```

```python
for snr_db in snr_values:
    args = {
        "mode": "predict_file",
        "pretrained_model": "nisqa.tar",
        "deg": f"../audiofiles/noisy_{snr_db}db.wav",
        "ms_channel": None,
        "output_dir": None,
    }
    model = nisqaModel(args)
    results = model.predict()

for snr_db in snr_values:
    args = {
        "mode": "predict_file",
        "pretrained_model": "nisqa.tar",
        "deg": f"../audiofiles/denoised_{snr_db}db.wav",
        "ms_channel": None,
        "output_dir": None,
    }
    model = nisqaModel(args)
    results = model.predict()
```
